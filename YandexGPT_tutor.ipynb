```python
import os
import openai

YANDEX_CLOUD_FOLDER = "b1g3srca2crvgvuamaud"
YANDEX_CLOUD_API_KEY = os.environ['yandexgpt-api-key']
YANDEX_CLOUD_MODEL = "yandexgpt-lite" # yandexgpt-lite or yandexgpt

```

```python

client = openai.OpenAI(
    api_key=YANDEX_CLOUD_API_KEY,
    base_url="https://rest-assistant.api.cloud.yandex.net/v1",
    project= YANDEX_CLOUD_FOLDER
)

response = client.responses.create(
    model=f"gpt://{YANDEX_CLOUD_FOLDER}/{YANDEX_CLOUD_MODEL}",
    input="Придумай 3 необычные идеи для стартапа в сфере путешествий.",
    temperature=0.8,
    max_output_tokens=1500
)

print(response.output[0].content[0].text)
```
